```python
def add_product(products, name, price):
    for product in products:
        if product["name"].lower() == name.lower():
            print("Error: Duplicate product name.")
            return
    products.append({"name": name, "price": float(price)})
    print("Product added successfully.")


def delete_product(products, name):
    for product in products:
        if product["name"].lower() == name.lower():
            products.remove(product)
            print("Product deleted successfully.")
            return
    print("Error: Product not found.")


def edit_product(products, name, new_name, new_price):
    for product in products:
        if product["name"].lower() == name.lower():
            product["name"] = new_name
            product["price"] = float(new_price)
            print("Product updated successfully.")
            return
    print("Error: Product not found.")


def search_product(products, name):
    for product in products:
        if product["name"].lower() == name.lower():
            return product
    return None


def show_product_details(products):
    if len(products) == 0:
        print("No products found.")
        return

    total_products = len(products)
    total_price = sum([product["price"] for product in products])
    avg_price = total_price / total_products
    sorted_products = sorted(products, key=lambda x: x["price"])
    cheapest_product = sorted_products[0]
    most_expensive_product = sorted_products[-1]

    print(f"Total number of products: {total_products}")
    print(f"Average price: ${avg_price:.2f}")
    print(f"Cheapest product: {cheapest_product['name']} (${cheapest_product['price']:.2f})")
    print(f"Most expensive product: {most_expensive_product['name']} (${most_expensive_product['price']:.2f})")


def main():
    products = []

    while True:
        print("\nMenu:")
        print("1. Add product")
        print("2. Delete product")
        print("3. Edit product")
        print("4. Search product")
        print("5. Show product details")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            name = input("Enter product name: ")
            price = float(input("Enter product price: "))
            add_product(products, name, price)
        elif choice == "2":
            name = input("Enter product name to delete: ")
            delete_product(products, name)
        elif choice == "3":

            name = input("Enter product name to edit: ")
            new_name = input("Enter new product name: ")
            new_price = float(input("Enter new product price: "))
            edit_product(products, name, new_name, new_price)
        elif choice == "4":
            name = input("Enter product name to search: ")
            product = search_product(products, name)
            if product:
                print(f"Product found: {product['name']} (${product['price']:.2f})")
            else:
                print("Product not found.")
        elif choice == "5":
            show_product_details(products)
        elif choice == "6" or choice.lower() == "exit":
            break
        else:
            print("Invalid input. Please try again.")


```

```python
main()
```
